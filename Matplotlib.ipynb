# Matplotlib

It gives you control over every aspect of a figure or a plot. designed to have a similar feel to MATLAB's graphcal plotting
matplotlib works very well with pandas and numpy arrays. website: #matplotlib.org use the site.


```python
import matplotlib.pyplot as plt
%matplotlib inline 
#for us to see plots inside jupyter notebook
```

if you are not using jupyter type plt.show() after all ur plots/code

```python
import numpy as np
x= np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#basic matplotlib plots, 2 ways of creating functional and object oriented(better way)
#later we will learn how to specify dpi(dots per inch) in order to get some more quality such as a finer font
#functional plotting
plt.plot(x,y,'r-') #colorstyles or line styles later
plt.xlabel('x')
plt.ylabel('y')
plt.title('My first plot')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

now we will break this all down for a more formal introduction to matplotlibs object oriented API method. This means we are going to instantiate figure objects and call methods or attributes from that object
OBJECT ORIENTED METHOD FOLLOWS.

```python
fig2= plt.figure()
axes= fig2.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('y')
axes.set_title('chart using ooc')
```

```python
fig= plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.5, 0.5,0.2,0.2])
axes1.plot(x,y)
axes1.set_title('big')
axes2.plot(y,x)
axes2.set_title('small')
```

```python
fig,axes= plt.subplots(nrows=1, ncols=2)
#plt.tight_layout()
for i in axes:
    i.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('hello world')
axes[1].plot(y,x)
plt.tight_layout()
```

# figure size aspect ratio and dpi( dots per inch or pixels per inch)


```python
fig=plt.figure(figsize=(1,1), dpi= 100)
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig1, axes = plt.subplots(nrows=2, ncols=1, figsize= (8,2))
for i in axes:
    i.plot(y,x)
plt.tight_layout()
```

```python
fig1.savefig('pic.png', dpi= 200)
```

```python
gig = plt.figure()
ax= gig.add_axes([0,0,1,1])
ax.plot(x,y, label= 'x squared')
ax.plot(x,x**3, label= 'x cubed')
ax.legend(loc= (0.1,0.1))
ax.legend(loc= 0)
```

```python
new= plt.figure()
axe=new.add_axes([0,0,1,1])
axe.plot(y,x, color= 'red',linewidth=5, alpha=0.5)
axe.plot(x,y, color= 'purple',linewidth=5, alpha= 0.5)
```
